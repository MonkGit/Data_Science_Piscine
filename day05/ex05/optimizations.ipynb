loop: write a function that iterates through the dataframe using for i in range(0, len(df)), iloc and append() to a list, 

assign the result of the function to a new column in the dataframe

```python
%%time
new_col = []

for i in range(0, len(df)):
    new_col.append(df.iloc[i]['Fines'] / (df.iloc[i]['Refund'] * df.iloc[i]['Year']))

df['Fines / Refund * Year'] = new_col
```

do it using iterrows()

```python
%%time
new_col = []

for index, row in df.iterrows():
    new_col.append(row['Fines'] / (row['Refund'] * row['Year']))

df['Fines / Refund * Year'] = new_col
```

do it using apply() and lambda function

```python
%%time
df['Fines / Refund * Year'] = df.apply(
    lambda row: row['Fines'] / (row['Refund'] * row['Year']), axis=1
    )
```

do it using Series objects from the dataframe

```python
%%time
df['Fines / Refund * Year'] = df['Fines'] / (df['Refund'] * df['Year'])
```

do it as in the previous subtask but with the method .values

```python
%%time
df['Fines / Refund * Year'] = df['Fines'].values / (
    df['Refund'].values * df['Year'].values
    )
```
